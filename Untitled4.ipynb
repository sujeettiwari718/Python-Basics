```python
#Question 2 : Explain the difference between List and Tuple in Python.

#ANSWER 2 :
 Difference Between List and Tuple in Python
  List :- A List is a collection of items that can be changed (mutable).
          List uses square brackets [ ].
          Items can be added, removed, or modified.
          Lists are generally slower.

  Tuple :- A Tuple is a collection of items that cannot be changed (immutable).
           Tuple uses parentheses ( ).
           Items cannot be added, removed, or modified after creation.
           Tuples are generally faster than lists.
```

```python
#Question 3 : What is a function in Python? Why are functions useful?

#ANSWER 3: 
        What is a Function in Python?

          A function in Python is a block of reusable code that performs a specific task.
          Functions help organize code and allow us to use the same code multiple times without rewriting it.

        Why are Functions Useful?

           1. Code Reusability
              A function can be used many times, reducing repeated code.
           2. Better Code Organization
              Functions make programs easier to read, understand, and manage.
           3. Easy Debugging and Maintenance
              Errors can be found and fixed easily because code is divided into smaller parts.
           4. Saves Time and Effort
              We write the code once and call it whenever needed.
```

```python
#Question 5 : Write a Python program to check whether a number is even or odd using conditional statements.

#ANSWER 5: 
number = int(input("Enter a number: "))

if number % 2 == 0:
    print("The number is Even")
else:
    print("The number is Odd")
```

```python
#Question 8 : Write a Python program to remove duplicate values from a list using a set.

#ANSWER 8:
numbers = [1, 2, 3, 2, 4, 5, 1, 3]

unique_numbers = list(set(numbers))

print("List after removing duplicates:", unique_numbers)
```

```python
#Question 9 : Write a function that returns the square of a number.

#ANSWER 9:
def square(num):
    return num * num

result = square(5)
print("Square is:", result)
```

```python
#Question 10 : Write a Python program to count how many times a number appears in a list. Example List [2,3,4,2,5,2]

#ANSWER 10: 
numbers = [2, 3, 4, 2, 5, 2]

count = numbers.count(2)

print("2 appears", count, "times in the list")
```
